```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
print(df.isnull().sum())
print('Total : ' + str(df.isnull().sum().sum()))
```

```python
df.fillna(df.select_dtypes(include=np.number).mean(), inplace=True)
print(df.isnull().sum())
print('Total : ' + str(df.isnull().sum().sum()))
```

```python
df.fillna(df.select_dtypes(include=np.number).median(), inplace=True)
print(df.isnull().sum())
print('Total : ' + str(df.isnull().sum().sum()))
```

```python
df.fillna(df.select_dtypes(include=np.number).mode(), inplace=True)
print(df.isnull().sum())
print('Total : ' + str(df.isnull().sum().sum()))
```

```python
df.dropna(inplace=True)
print(df.isnull().sum())
print('Total : ' + str(df.isnull().sum().sum()))
```

```python
df_before = df.copy()
df_before.hist(figsize=(10,8), bins=20)
plt.suptitle("Histograms before filling missing values")
plt.tight_layout()
plt.show()

df_after = df_before.fillna(df_before.mean(numeric_only=True))
df_after.hist(figsize=(10,8), bins=20)
plt.suptitle("Histograms after filling missing values:")
plt.tight_layout()
plt.show()
```
